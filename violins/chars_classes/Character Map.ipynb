```python
from collections import defaultdict, Counter
import pandas

def to_int(x, y):
    return int(x), y

def get_milestones(text):
    with open(text) as f:
        milestones = dict([
            to_int(*s.strip().split("#"))
            for s in f
        ])
    return milestones
```

## TNAH

```python
stud_mls = get_milestones("tnah.milestones.txt")
stud_txt = defaultdict(str)
full_txt = ""

with open("training_students.txt") as f:
    current_text = None
    for no, line in enumerate(f):
        if no in stud_mls:
            current_text = stud_mls[no]
        stud_txt[current_text] += line.strip()
        stud_txt["Full"] += line.strip()
        


counters = {
    title: Counter(text)
    for title, text in stud_txt.items()
}
majs = {
    title: Counter([
        el
        for key, value in cnt.items()
        for el in (
            value * ["Uppercase"]
            if key.isupper() 
            else 
            value * ["Lowercase"]
        )
    ])
    for title, cnt in counters.items()
}
    
diversity = [
    [
        title,
        dict([
            ("size", len(text)), 
            ("dataset", "students"), 
            ("classes", len(set(text))), 
            *majs[title].items(), 
            *counters[title].items()
        ])
    ]
    for title, text in stud_txt.items()
]
```

## Pinche

```python
pinc_mls = get_milestones("pinche.milestones.txt")
pinc_txt = defaultdict(str)

with open("training full.txt") as f:
    current_text = None
    for no, line in enumerate(f):
        if no in pinc_mls:
            current_text = pinc_mls[no]
        pinc_txt[current_text] += line.strip()
        pinc_txt["Full"] += line.strip()
        


counters = {
    title: Counter(text)
    for title, text in pinc_txt.items()
}
majs = {
    title: Counter([
        el
        for key, value in cnt.items()
        for el in (
            value * ["Uppercase"]
            if key.isupper() 
            else 
            value * ["Lowercase"]
        )
    ])
    for title, cnt in counters.items()
}


diversity.extend([
    *[
        [
            title,
            dict([
                ("size", len(text)), 
                ("dataset", "pinche"), 
                ("classes", len(set(text))), 
                *majs[title].items(), 
                *counters[title].items()
            ])
        ]
        for title, text in pinc_txt.items()
    ]
])
```

## Dataframe

```python
df = pandas.DataFrame([d[1] for d in diversity], index=[d[0] for d in diversity]).fillna(0)
df.columns = [
    5*[""]+(len(df.columns)-5)*["Char"],
    df.columns
]
df.to_csv("datasets-characters.csv")
#df["Char"].div(df[("", "size")], axis=0)
df
```
